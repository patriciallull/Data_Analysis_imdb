## Question 7
What's the average value of runtimeMinutes for the type short?

```python
short_movies = imdb[imdb['titleType']=='short']
average_runtime= round(pd.Series.mean(short_movies['runtimeMinutes'], skipna= True),1) #the default is skipna= True just add it here to remember
print(f"the average runtime for short movies released in 2019 is: {average_runtime} minutes")
```
